```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import Linear_Regression as lr # Custom created module implements Linear regression using Gradient Descent
```

# Validate Models
## Momentum

```python
# Toy data
x = np.linspace(0, 20)
y = -2 * x + 1

# Test Toy data on model, the points and line should coincide
X = np.vstack((np.ones_like(x), x)).T
y = np.atleast_2d(y).T
lr_test_momentum = lr.Linear_regression(X, y)
lr_test_momentum.fit(solver="momentum", stop_criteria=1e-4)
lr_test_momentum.show_summary()
lr_test_momentum.plot_LR_2D()
lr_test_momentum.plot_MSE()
```

## Nesterov

```python
lr_test_NAG = lr.Linear_regression(X, y)
lr_test_NAG.fit(solver="Nesterov", stop_criteria=1e-4)
lr_test_NAG.show_summary()
lr_test_NAG.plot_LR_2D()
lr_test_NAG.plot_MSE()
```

# Noisy data

```python
# Read Data
df = pd.read_csv('RegData.csv', header=None)
df.head()
```

```python
# Prepare data as numpy arrays
X = np.atleast_2d(df[0].to_numpy()).T
y = df[1].to_numpy()
# Add a row of zeros for theta_0
X = np.hstack((np.ones((len(X), 1)), X))
y = np.atleast_2d(y).T
```

# Momentum

```python
# Initialize object from class Linear regression and solve
# Momentum GD
lr_momentum = lr.Linear_regression(X, y)
theta = lr_momentum.fit(solver="momentum", alpha=1e-3, max_epochs=1e3)
lr_momentum.show_summary()
lr_momentum.plot_LR_2D(show_trials=True)
lr_momentum.plot_MSE()
```

## Change values of $\gamma$ in momentum

```python
# Initialize object from class Linear regression and solve
# Momentum GD
lr_momentum = lr.Linear_regression(X, y)
theta = lr_momentum.fit(solver="momentum", alpha=1e-3, max_epochs=1e3, gamma=0.2)
lr_momentum.show_summary()
lr_momentum.plot_LR_2D(show_trials=True)
lr_momentum.plot_MSE()
```
