Building a PCA (Principal Component Analysis) implementation on Sklearn Iris data set.

```python
import numpy as np
from numpy.linalg import eig
from sklearn import datasets
import matplotlib.pyplot as plt
```

# Homework 4

```python
#PCA algorithm
def PCA(A, n_components):
    """
    Function to find the Principal components of dataset A
    
    Parameters:
    ==========
    A: Design matrix of a shape (m, n);
    m -> Each row is a new obsevartion
    n -> Each column is a new feature
    
    n_components: Number of components to return
    
    returns:
    =======
    P: Principal components matrix, with "n_components". 
    shape: (m, n_components)
    """
    means = np.apply_along_axis(np.mean, 0, A)
    A_centrd = A - means
    
    cov_mat = np.cov(A_centrd.T)
    l, v = eig(cov_mat)
    P = A_centrd @ v[:, :n_components]
    
    return P
```

```python
#Load dataset from sklearn
iris = datasets.load_iris()
X = iris.data #Design Matrix
y = iris.target
```

```python
#Perform Principal component analysis
X_reduced = PCA(X, 2)

#Plot the results
fig, ax = plt.subplots()
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.Set1, edgecolor='k', s=45)
ax.set_title("First two PCA directions")
ax.set_xlabel("1st eigenvector")
ax.set_ylabel("2nd eigenvector")
plt.show()
```
